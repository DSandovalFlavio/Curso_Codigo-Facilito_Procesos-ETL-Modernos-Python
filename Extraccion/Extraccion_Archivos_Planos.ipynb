# Extraccion de datos desde archivos

- CSV
- Excel

```python
import pandas as pd
import polars as pl
import os

path_data = os.path.join(os.getcwd(), 'data') 
```

```python
path_data
```

## CSV

```python
pd.read_csv(
    filepath_or_buffer=os.path.join(path_data, 'dataset.csv'),
    )
```

```python
pd.read_csv(
    filepath_or_buffer=os.path.join(path_data, 'dataset.csv'),
    skiprows=1,
    delimiter='/',
    header=None,
    skipfooter=1, # se usa para saltar las ultimas filas
    names=['A', 'B', 'C'],
    )
```

```python
pd.read_csv(
    filepath_or_buffer=os.path.join(path_data, 'dataset.txt'),
    delimiter='\t'
    )
```

## Excel

```python
pd.read_excel(
    io=os.path.join(path_data, 'dataset_tipo2.xlsx'),
    )
```

```python
xls = pd.ExcelFile(os.path.join(path_data, 'datos_historicos.xlsx'))
xls.sheet_names
```

```python
df_full = pd.concat([pd.read_excel(xls, sheet_name=sheet) for sheet in xls.sheet_names], ignore_index=True)
```
